```python
import pandas as pd

df = pd.read_csv("Earthquakes_database.csv")
```

```python
df
```

```python
df[['Month', 'Day', 'Year']] = df['Date'].str.split('/', expand=True)

df[['Hour', 'Minute', 'Second']] = df['Time'].str.split(':', expand=True)
```

```python
df
```

```python
df = df.dropna(subset=['Year'])

bins = [0, 1970, 2000, 2025]
labels = ['Pre-1970', '1970-2000', 'Post-2000']

df['Year_Category'] = pd.cut(df['Year'], bins=bins, labels=labels, right=True)
```

```python
df
```

```python
df['Year'] = df['Year'].astype(int)


melted_df = pd.melt(df, id_vars=['Date', 'Type'], value_vars=['Latitude', 'Longitude', 'Depth', 'Magnitude'])

print("\nDataFrame after melting:")
print(melted_df.head())

pivot_table_df = pd.pivot_table(df, values=['Depth', 'Magnitude'], index=['Year', 'Type'], aggfunc='mean')

print("\nPivot Table:")
print(pivot_table_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Year'] = df['Year'].astype(int)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Year')
plt.title('Number of Earthquakes Over the Years')
plt.xlabel('Year')
plt.ylabel('Count of Earthquakes')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Year_Category', y='Magnitude')
plt.title('Average Magnitude by Year Category')
plt.xlabel('Year Category')
plt.ylabel('Average Magnitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Depth', y='Magnitude')
plt.title('Depth vs Magnitude')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.show()
```

```python
type_counts = df['Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Earthquake Types')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Magnitude', bins=20, kde=True)
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python

data = {
    'Date': ['1/2/1965', '1/4/1965', '1/5/1965', '1/8/1965', '1/9/1965'],
    'Magnitude': [6.0, 5.8, 6.2, 5.8, 5.8],
    'Status': ['Automatic', 'Automatic', 'Automatic', 'Reviewed', 'Reviewed'],
    'Latitude': [19.2460, 1.8630, -20.5790, -59.0760, 11.9380],
    'Longitude': [145.6160, 127.3520, -173.9720, -23.5570, 126.4270],
}

df = pd.DataFrame(data)

```

```python
print(df.columns)
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

magnitude_weekly_mean = df.resample('W')['Magnitude'].mean()
weekly_summary = df.resample('W').agg({
    'Magnitude': 'sum',
    'Latitude': 'max',
    'Longitude': 'min'
})

print("Weekly mean magnitude:")
print(magnitude_weekly_mean)
print("\nWeekly summary:")
print(weekly_summary)
```
